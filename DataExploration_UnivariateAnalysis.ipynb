(3) In which course are students more successful in your chosen state? Mathematics or reading? (Seçtiğiniz eyalette öğrenciler hangi derste daha başarılı? Matematik mi yoksa okuma mı?)

AVG_MATH_4_SCORE: The state's average score for fourth graders taking the NAEP math exam.

AVG_MATH_8_SCORE: The state's average score for eight graders taking the NAEP math exam.

AVG_READING_4_SCORE: The state's average score for fourth graders taking the NAEP reading exam.

AVG_READING_8_SCORE: The state's average score for eighth graders taking the NAEP reading exam.

Hem 4. sınıf hem de 8. sınıf öğrencileri bulunduğu için bunların ortalamasını alacağız.

```python
math_4 = california.AVG_MATH_4_SCORE.mean()
math_8 = california.AVG_MATH_8_SCORE.mean()
ort_math = (math_4 + math_8)/2
ort_math
```

```python
reading_4 = california.AVG_READING_4_SCORE.mean()
reading_8 = california.AVG_READING_8_SCORE.mean()
ort_reading= (reading_4+reading_8)/2
ort_reading
```

Görüldüğü üzere Matematik ortalaması reading ortalamasından yüksek. Matematik dersinde daha başarılılarmış. Aşağıda yeni feature oluşturup bunlara describe ile bakacağım.

```python
california['AVARAGE_MATH'] = (california['AVG_MATH_4_SCORE'] + california['AVG_MATH_8_SCORE'])/2
```

```python
california['AVARAGE_READING'] = (california['AVG_READING_4_SCORE'] + california['AVG_READING_8_SCORE'])/2
```

(4) What is the distribution of mathematics and reading grades?(Matematik ve okuma notlarının dağılımı nasıldır?)

```python
plt.figure(figsize=(10, 5), dpi = 100)

plt.subplot(1,2,1)
plt.hist(california.AVARAGE_MATH, density=True, color='red', alpha=.5, label ="math") 
plt.legend()
plt.subplot(1,2,2)
plt.hist(california.AVARAGE_READING, density=True, color='green', alpha=.5 ,label="reading")

plt.legend()
plt.show()
```
